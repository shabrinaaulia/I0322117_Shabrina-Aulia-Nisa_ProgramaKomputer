```python
#Buatlah Fungsi untuk melakukan pembagian matrix antar matrix berdasarkan input user menggunakan nested loop

def bagi_matrix(matrix1, skalar):
    hasil = []

    #Tulis kode kamu disini
    for i in range(len(matrix1)) :
        row =[]
        for j in range(len(matrix1[0])) :
            elemen = matrix1[i][j] / skalar
            row.append(elemen)
        hasil.append(row)


    return hasil


matrix1 = []

# Menginput elemen matrix pertama
baris = int(input("Masukkan jumlah baris"))
kolom = int(input("Masukkan jumlah kolom"))

for i in range(baris):
    row = []
    for j in range(kolom):
        elemen = int(input(f"Masukkan elemen matrix pertama [{i}][{j}]: "))
        row.append(elemen)
    matrix1.apppend(row)

skalar = int(input('Matrix dibagi berapa?'))

hasil_pembagian = bagi_matrix(matrix1,skalar)
print("Hasil pembagian matrix : ")
for row in hasil_pembagian:
    print(row)
```
